## What deck were the passengers on and how does that relate to their classs

```python
titanic_df.head()
```

```python
titanic_df.head(2)
```

```python
deck = titanic_df['Cabin'].dropna()
deck
```

```python
levels = []
for level in deck:
  levels.append(level[0])

cabin_df = pd.DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df.sort_values(by='Cabin', inplace=True)
sns.catplot(x='Cabin', data=cabin_df, kind='count', palette='pastel')
```

```python
cabin_df
sns.catplot(x='Cabin', data=cabin_df, kind='count', palette='pastel')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.catplot(x='Cabin', data=cabin_df, kind='count', palette='summer')
```

# **Who was with family and who was alone ?**

```python
titanic_df['Alone']=titanic_df.SibSp+titanic_df.Parch
titanic_df['Alone']
```

```python
titanic_df['Alone'].value_counts()
```

```python
titanic_df.loc[titanic_df['Alone']>0 , 'Alone']='With Family'
titanic_df.loc[titanic_df['Alone']==0 , 'Alone']='Alone'
```

```python
titanic_df.head()
```

```python
sns.catplot(x='Alone',data=titanic_df,kind='count',palette='pastel',order=['Alone','With Family'])
```
