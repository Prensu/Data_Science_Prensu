## **Titanic Dataset**


```python
import pandas as pd
url='https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
titanic_df=pd.read_csv(url)
titanic_df
```

```python
print(f"How much missing value in every column: {titanic_df.isna().sum()}")
```

**Who were the passengers in the titanic ?**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('ticks')
plt.style.use('fivethirtyeight')
```

```python
#catplot-seperates out gender
sns.catplot(x='Sex',kind='count',data=titanic_df)
```

```python
sns.catplot(x='Sex',kind='count',data=titanic_df,palette={'male':'red','female':'pink'})
plt.title("Count of passengers by gender")
plt.show()
```

```python
sns.catplot(x='Pclass',kind='count',data=titanic_df,hue='Sex')
plt.title("Count of passengers by class")
plt.show()
```

Create a new column 'Person' in which every person under 16 is child **

```python
titanic_df['Person']=titanic_df.Sex
titanic_df.loc[titanic_df['Age']<16 , 'Person']='Child'
titanic_df
```
