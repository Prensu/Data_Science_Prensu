```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5]
y=[10,15,25,30,20]

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.plot(x,y,marker='o',linestyle='--',color='green',label='myline')
plt.legend()
plt.show()
```
