```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('data'):
    with open(os.path.join('data', filename), 'r') as f:
        corpus = f.read() #from gameofthrones the data of first file is taken and stored in the corpus 
        raw_sent = sent_tokenize(corpus) #then using nltk library the raw data is sentence tokenized 
        for sent in raw_sent:  #taken every sentence and sent into simple preprocess function 
            story.append(simple_preprocess(sent))  #now everything preprocessing is done!and this is done for all 5 files 
```

```python
len(story)
```

```python
story #here the sentences are also tokenized into words
```

```python
model = gensim.models.Word2Vec(   # here i am just creating the model 
    window=10,  # we will take 10words left and 10 words right sliding we do 
    min_count=2, # we will consider the sentences with min 2 words in it 
)
```

```python
model.build_vocab(story) # in my corpus i have many unique words so i will be building vocabulary(part of training)
```

```python
# i had the whole corpus in the story so i sent all of them 
```

```python
model.train(story,total_examples=model.corpus_count ,epochs=model.epochs) #using .train to train my deep learning model 
```

```python
# model.corpus_count => will automatically store the number of sentences in the corpus during initialization 
```
