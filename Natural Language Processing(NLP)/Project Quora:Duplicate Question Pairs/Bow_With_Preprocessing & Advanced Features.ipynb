```python
def common_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))
    return len(w1 & w2)

```

```python
def total_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))
    return (len(w1) + len(w2))

new_df['word_total'] = new_df.apply(total_words, axis=1)
new_df.head()
```

```python
# Import the stopwords corpus from NLTK (Natural Language Toolkit)
from nltk.corpus import stopwords

# Define the function to fetch token-based features for a pair of questions
def fetch_token_features(row):
    """
    This function calculates token-based features for two questions (question1 and question2).
    It takes a row from a pandas DataFrame as input and returns a list of 8 features.
    """
    
    # Step 1: Extract the two questions from the row
    q1 = row['question1']  # Extract question1
    q2 = row['question2']  # Extract question2
    
    # Step 2: Define a small constant to avoid division by zero errors
    SAFE_DIV = 0.0001  # A small value added to denominators to prevent division by zero
    
    # Step 3: Load the list of English stopwords from NLTK
    STOP_WORDS = stopwords.words("english")  # List of common stopwords like "the", "is", "and", etc.
    
    # Step 4: Initialize a list to store the 8 token-based features
    token_features = [0.0] * 8  # Create a list with 8 zeros to store the calculated features
    
    # Step 5: Split the questions into tokens (words)
    q1_tokens = q1.split()  # Split question1 into words
    q2_tokens = q2.split()  # Split question2 into words
    
    # Step 6: Handle edge case where one of the questions has no tokens
    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return token_features  # If either question is empty, return the list of zeros
    
    # Step 7: Get the non-stopwords in each question
    q1_words = set([word for word in q1_tokens if word not in STOP_WORDS])  # Non-stopwords in q1
    q2_words = set([word for word in q2_tokens if word not in STOP_WORDS])  # Non-stopwords in q2
    
    # Step 8: Get the stopwords in each question
    q1_stops = set([word for word in q1_tokens if word in STOP_WORDS])  # Stopwords in q1
    q2_stops = set([word for word in q2_tokens if word in STOP_WORDS])  # Stopwords in q2
    
    # Step 9: Calculate the number of common non-stopwords between the two questions
    common_word_count = len(q1_words.intersection(q2_words))  # Count of common non-stopwords
    
    # Step 10: Calculate the number of common stopwords between the two questions
    common_stop_count = len(q1_stops.intersection(q2_stops))  # Count of common stopwords
    
    # Step 11: Calculate the number of common tokens (words) between the two questions
    common_token_count = len(set(q1_tokens).intersection(set(q2_tokens)))  # Count of common tokens
    
    # Step 12: Calculate the ratio of common non-stopwords to the minimum number of non-stopwords
    token_features[0] = common_word_count / (min(len(q1_words), len(q2_words)) + SAFE_DIV)
    
    # Step 13: Calculate the ratio of common non-stopwords to the maximum number of non-stopwords
    token_features[1] = common_word_count / (max(len(q1_words), len(q2_words)) + SAFE_DIV)
    
    # Step 14: Calculate the ratio of common stopwords to the minimum number of stopwords
    token_features[2] = common_stop_count / (min(len(q1_stops), len(q2_stops)) + SAFE_DIV)
    
    # Step 15: Calculate the ratio of common stopwords to the maximum number of stopwords
    token_features[3] = common_stop_count / (max(len(q1_stops), len(q2_stops)) + SAFE_DIV)
    
    # Step 16: Calculate the ratio of common tokens to the minimum number of tokens
    token_features[4] = common_token_count / (min(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    
    # Step 17: Calculate the ratio of common tokens to the maximum number of tokens
    token_features[5] = common_token_count / (max(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    
    # Step 18: Check if the last word of both questions is the same
    token_features[6] = int(q1_tokens[-1] == q2_tokens[-1])  # 1 if same, 0 if different
    
    # Step 19: Check if the first word of both questions is the same
    token_features[7] = int(q1_tokens[0] == q2_tokens[0])  # 1 if same, 0 if different
    
    # Step 20: Return the list of calculated features
    return token_features
```

```python
token_features = new_df.apply(fetch_token_features, axis=1)

new_df["cwc_min"]       = list(map(lambda x: x[0], token_features))
new_df["cwc_max"]       = list(map(lambda x: x[1], token_features))
new_df["csc_min"]       = list(map(lambda x: x[2], token_features))
new_df["csc_max"]       = list(map(lambda x: x[3], token_features))
new_df["ctc_min"]       = list(map(lambda x: x[4], token_features))
new_df["ctc_max"]       = list(map(lambda x: x[5], token_features))
new_df["last_word_eq"]  = list(map(lambda x: x[6], token_features))
new_df["first_word_eq"] = list(map(lambda x: x[7], token_features))
```

```python
import distance

def fetch_length_features(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    length_features = [0.0]*3
    
    # Converting the Sentence into Tokens: 
    q1_tokens = q1.split()
    q2_tokens = q2.split()
    
    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return length_features
    
    # Absolute length features
    length_features[0] = abs(len(q1_tokens) - len(q2_tokens))
    
    #Average Token Length of both Questions
    length_features[1] = (len(q1_tokens) + len(q2_tokens))/2
    
    strs = list(distance.lcsubstrings(q1, q2))
    length_features[2] = len(strs[0]) / (min(len(q1), len(q2)) + 1)
    
    return length_features
```

```python
length_features = new_df.apply(fetch_length_features, axis=1)

new_df['abs_len_diff'] = list(map(lambda x: x[0], length_features))
new_df['mean_len'] = list(map(lambda x: x[1], length_features))
new_df['longest_substr_ratio'] = list(map(lambda x: x[2], length_features))
```

```python
# Fuzzy Features
from fuzzywuzzy import fuzz

def fetch_fuzzy_features(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    fuzzy_features = [0.0]*4
    
    # fuzz_ratio
    fuzzy_features[0] = fuzz.QRatio(q1, q2)

    # fuzz_partial_ratio
    fuzzy_features[1] = fuzz.partial_ratio(q1, q2)

    # token_sort_ratio
    fuzzy_features[2] = fuzz.token_sort_ratio(q1, q2)

    # token_set_ratio
    fuzzy_features[3] = fuzz.token_set_ratio(q1, q2)

    return fuzzy_features
```

```python
fuzzy_features = new_df.apply(fetch_fuzzy_features, axis=1)

# Creating new feature columns for fuzzy features
new_df['fuzz_ratio'] = list(map(lambda x: x[0], fuzzy_features))
new_df['fuzz_partial_ratio'] = list(map(lambda x: x[1], fuzzy_features))
new_df['token_sort_ratio'] = list(map(lambda x: x[2], fuzzy_features))
new_df['token_set_ratio'] = list(map(lambda x: x[3], fuzzy_features))
```
