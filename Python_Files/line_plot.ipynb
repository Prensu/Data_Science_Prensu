```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5]
y=[10,15,25,30,20]

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.plot(x,y,marker='o',linestyle='--',color='green',label='myline')
plt.legend()
plt.show()
```

```python
x2=np.array([1,2,3,4,5])
y2=np.array([10,15,25,30,20])

plt.plot(x2,y2,marker='o',linestyle='--',color='green',label='circle')

plt.plot(x2,y2,marker='s',linestyle='--',color='red',label='square')

plt.plot(x2,y2,marker='^',linestyle='--',color='blue',label='triangle')

plt.plot(x2,y2,marker='*',linestyle='--',color='black',label='star')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot"	)
plt.legend()
plt.show()
```

Multiple line in a single plot



```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5]
y=[10,15,25,30,20]
y2=[5,10,15,20,25]
plt.plot(x,y,label='line1')
plt.plot(x,y2,linestyle='--',color='green',label='Dashed Red Line')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in a single plot")
plt.legend()
plt.show()
```

Labelled plot with positioned legends !

```python
plt.plot(x,y,label='line1')
plt.plot(x,y2,label='line2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("labelled plot with positioned legends ")
plt.legend(loc='upper left',fontsize=12,frameon=False)
plt.show()
```

With Gridlines

```python
plt.plot(x,y,label='line1')
plt.plot(x,y2,label='line2')
plt.grid()
plt.legend()
plt.show()
```

More on GridLines

```python
plt.subplots(nrows=2,ncols=2,num=1)
plt.subplots(nrows=2,ncols=2,num=2)
plt.subplots(nrows=2,ncols=2,num=3)
plt.subplots(nrows=2,ncols=2,num=4)
plt.plot(x,y,label='line1')
plt.plot(x,y2,label='line2')
plt.plot(x,y,label='line1')
plt.plot(x,y2,label='line2')
plt.plot(x,y,label='line1')
plt.plot(x,y2,label='line2')
plt.grid(color='gray',linestyle='--',linewidth=0.5)
plt.legend()
plt.show()
```

# Using Random Data using numpy and  **rand**

```python
import numpy as np
x=np.random.normal(loc=170,scale=10,size=250)
plt.hist(x)
plt.show()
```
