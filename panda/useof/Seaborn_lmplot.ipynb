```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, ci = 75)
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue = "time", 
           markers = ['d', 'x'], palette =["blue", 'red']
          )
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue = "time", 
           markers = ['d', 'x'], palette =["blue", 'red'],
           col = "day", row = 'sex',
           height = 4, aspect = 0.7
          )
```

```python
sns.lmplot(x = "size", y = "tip", data = tips, hue = "time", 
           markers = ['d', 'x'], palette =["blue", 'red'],
           col = "day", row = 'sex',
           height = 4, aspect = 0.7
          )
```

```python
sns.lmplot(x = "size", y = "tip", data = tips, hue = "time", 
           markers = ['d', 'x'], palette =["blue", 'red'],
           col = "day", row = 'sex', x_jitter = 0.2,
           height = 4, aspect = 0.7
          )
```

```python
p = sns.lmplot(x = "size", y = "tip", data = tips, hue = "time", 
           markers = ['d', 'x'], palette =["blue", 'red'],
           col = "day", row = 'sex', x_jitter = 0.2,
           height = 4, aspect = 0.7
          )

p.set_axis_labels("Size", "Tip").set(xlim = (0, 7), ylim=(0, 10),
                                    xticks = [1, 2, 3, 4, 5, 6, 7], yticks = [4, 8, 10]).fig.subplots_adjust(hspace = 0.25, wspace = 0.2)
```
