```python
import cv2
import numpy as np

# Load the main image and the template image
main_image = cv2.imread('soccer_practice.jpg')  # Replace with your main image path
template_image = cv2.imread('ball.png', 0)  # Replace with your template image path

# Convert the main image to grayscale
main_gray = cv2.cvtColor(main_image, cv2.COLOR_BGR2GRAY)

# Perform template matching
result = cv2.matchTemplate(main_gray, template_image, cv2.TM_CCOEFF_NORMED)

# Get the best match location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Get the dimensions of the template
template_height, template_width = template_image.shape

# Draw a rectangle around the detected football
top_left = max_loc
bottom_right = (top_left[0] + template_width, top_left[1] + template_height)
cv2.rectangle(main_image, top_left, bottom_right, (0, 255, 0), 2)  # Green rectangle

# Display the result
cv2.imshow('Detected Football', main_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the main image and template in grayscale
img = cv2.imread('soccer_practice.jpg', 0)  # 0 for grayscale
template = cv2.imread('ball.png', 0)
h, w = template.shape

# List of template matching methods
methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR,
           cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]

# Loop through each method
for method in methods:
    img2 = img.copy()  # Create a copy of the original image for drawing rectangles

    # Perform template matching
    result = cv2.matchTemplate(img2, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # Determine the location of the best match
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        location = min_loc
    else:
        location = max_loc

    # Define the bottom-right corner of the rectangle
    bottom_right = (location[0] + w, location[1] + h)

    # Draw the rectangle on the image
    cv2.rectangle(img2, location, bottom_right, 255, 5)

    # Display the result
    cv2.imshow('Match', img2)

    # Wait for a key press
    key = cv2.waitKey(0) & 0xFF  # Mask to get the last 8 bits

    # Check if 'q' is pressed
    if key == ord('q'):  # ord('q') returns the ASCII value of 'q'
        break  # Exit the loop if 'q' is pressed

# Close all OpenCV windows
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the main image and template in grayscale
img = cv2.imread('soccer_practice.jpg', 0)  # Replace with your main image path
template = cv2.imread('shoe.png', 0)  # Replace with your template image path (shoe)

# Resize the images (adjust scale_percent as needed)
scale_percent = 50  # Resize scale (percentage of original size)
img = cv2.resize(img, (int(img.shape[1] * scale_percent / 100), int(img.shape[0] * scale_percent / 100)))
template = cv2.resize(template, (int(template.shape[1] * scale_percent / 100), int(template.shape[0] * scale_percent / 100)))

# Get new dimensions of the resized template
h, w = template.shape

# List of template matching methods
methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR,
           cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]

# Create a figure to display multiple plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 rows, 3 columns for 6 methods
fig.suptitle('Template Matching Results', fontsize=16)

# Loop through each method and plot results
for i, method in enumerate(methods):
    img2 = img.copy()  # Create a copy of the original image for drawing rectangles

    # Perform template matching
    result = cv2.matchTemplate(img2, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # Determine the location of the best match
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        location = min_loc
    else:
        location = max_loc

    # Define the bottom-right corner of the rectangle
    bottom_right = (location[0] + w, location[1] + h)

    # Draw the rectangle on the image
    cv2.rectangle(img2, location, bottom_right, 255, 5)

    # Plot the results in a grid
    ax = axes[i // 3, i % 3]  # Determine subplot position (2 rows, 3 columns)
    ax.imshow(img2, cmap='gray')  # Show image in grayscale
    ax.set_title(f'Method: {method}')  # Title with method name
    ax.axis('off')  # Hide axes

# Show all plots in one figure
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit titles
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the main image and template in grayscale
img = cv2.imread('soccer_practice.jpg', 0)  # Replace with your main image path
template = cv2.imread('shoe.png', 0)  # Replace with your template image path (shoe)

# Resize the images (adjust scale_percent as needed)
scale_percent = 50  # Resize scale (percentage of original size)
img = cv2.resize(img, (int(img.shape[1] * scale_percent / 100), int(img.shape[0] * scale_percent / 100)))
template = cv2.resize(template, (int(template.shape[1] * scale_percent / 100), int(template.shape[0] * scale_percent / 100)))

# Get new dimensions of the resized template
h, w = template.shape

# List of template matching methods and their names
methods = {
    "TM_CCOEFF": cv2.TM_CCOEFF,
    "TM_CCOEFF_NORMED": cv2.TM_CCOEFF_NORMED,
    "TM_CCORR": cv2.TM_CCORR,
    "TM_CCORR_NORMED": cv2.TM_CCORR_NORMED,
    "TM_SQDIFF": cv2.TM_SQDIFF,
    "TM_SQDIFF_NORMED": cv2.TM_SQDIFF_NORMED
}

# Create a figure to display multiple plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 rows, 3 columns for 6 methods
fig.suptitle('Template Matching Results', fontsize=16)

# Loop through each method and plot results
for i, (method_name, method) in enumerate(methods.items()):
    img2 = img.copy()  # Create a copy of the original image for drawing rectangles

    # Perform template matching
    result = cv2.matchTemplate(img2, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # Determine the location of the best match
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        location = min_loc
    else:
        location = max_loc

    # Define the bottom-right corner of the rectangle
    bottom_right = (location[0] + w, location[1] + h)

    # Draw the rectangle on the image
    cv2.rectangle(img2, location, bottom_right, 255, 5)

    # Plot the results in a grid
    ax = axes[i // 3, i % 3]  # Determine subplot position (2 rows, 3 columns)
    ax.imshow(img2, cmap='gray')  # Show image in grayscale
    ax.set_title(f'Method: {method_name}')  # Title with method name
    ax.axis('off')  # Hide axes

# Show all plots in one figure
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit titles
plt.show()
```
