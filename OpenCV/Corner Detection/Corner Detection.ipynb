```python
import cv2
import numpy as np  # Ensure numpy is imported for corner handling

# Read the image
img = cv2.imread('Chessboard.png')

# Resize the image
img = cv2.resize(img, (0, 0), fx=0.75, fy=0.75)

# Convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect corners using goodFeaturesToTrack
corners = cv2.goodFeaturesToTrack(img_gray, 100, 0.01, 10)

# Draw the corners on the image
if corners is not None:
    corners = np.int0(corners)  # Convert corners to integer type
    for corner in corners:
        x, y = corner.ravel()  #Flatten an array
        cv2.circle(img, (x, y), 5, (0, 255, 0), -1)  # Draw a circle at each corner

# Display the image with corners
cv2.imshow('Frame', img)
cv2.waitKey(0)  # Wait for a key press
cv2.destroyAllWindows()  # Close all OpenCV windows
```
